\begin{equation}
\frac{d^2y}{dt^2} + \gamma \frac{dy}{dt} + \omega² y =0
\end{equation}


\begin{equation}
\frac{\mathrm{d}y}{\mathrm{d}t} =v
\end{equation}


\begin{equation}
\frac{\mathrm{d}v}{\mathrm{d}t} = - \gamma v - \omega² y
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def f(A,t,roz,w):
  y=A[0]
  v=A[1]
  dy=v
  dv=-roz*v-y*w**2
  return np.array([dy,dv])

y0 = np.array([2, 0])
fr=2
omega = 2*np.pi*fr
alpha = 0.1
t0 = np.linspace(0, 10, 100000)
solutionED = odeint(f, y0, t0, args=(alpha,omega))

plt.plot(t0, solutionED[:,0])
```

```python
#Euler pendulo
def dev(v,y,w,rh):
  return -rh*v-y*w**2
def dey(v):
  return v
a,b,h=0,10,0.0001

N=int((b-a)/h)
t=np.linspace(a,b,N+1)
y=np.zeros(N+1)
v=np.zeros(N+1)

y[0],v[0]=2,0
for i in range(0,N):
  v[i+1]=v[i]+ dev(v[i],y[i],omega,alpha)*h
  y[i+1]=y[i]+ dey(v[i])*h


plt.plot(t,y)
```

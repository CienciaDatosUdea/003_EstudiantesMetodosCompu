```python
import numpy as np
import matplotlib.pyplot as plt

def g(y,k):
  return -k*y
a,b,h=0,3,0.01
N=int((b-a)/h)
t=np.linspace(a,b,N+1)
y=np.zeros(N+1)

y[0]=100
for i in range(0,N):
  y[i+1]=y[i]+ g(y[i],10)*h


y_real=100*np.exp(-10*t)


plt.plot(t,y, label="funcion con euler ")
plt.plot(t,y_real, label="funcion Real")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema de Lorenz
sigma = 10
rho = 28
beta = 8/3

# Condiciones iniciales
x0, y0, z0 = 1.0, 1.0, 1.0

# Parámetros de integración
dt = 0.01  # Paso de tiempo
t_max = 50  # Tiempo máximo de simulación
n = int(t_max / dt)  # Número de pasos

# Inicialización de las soluciones
x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)
t = np.linspace(0, t_max, n)

# Asignar condiciones iniciales
x[0], y[0], z[0] = x0, y0, z0

# Integración numérica con Euler
for i in range(n - 1):
    dx = sigma * (y[i] - x[i]) * dt
    dy = (x[i] * (rho - z[i]) - y[i]) * dt
    dz = (x[i] * y[i] - beta * z[i]) * dt

    x[i + 1] = x[i] + dx
    y[i + 1] = y[i] + dy
    z[i + 1] = z[i] + dz

# Gráficos de la solución
fig = plt.figure(figsize=(10, 5))

# Gráfica en 3D del atractor
ax = fig.add_subplot(121, projection="3d")
ax.plot(x, y, z, color="b")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Atractor de Lorenz")

# Gráfica 2D del tiempo vs x
plt.subplot(122)
plt.plot(t, x, label="x(t)", color="r")
plt.plot(t, y, label="y(t)", color="g")
plt.plot(t, z, label="z(t)", color="b")
plt.xlabel("Tiempo")
plt.ylabel("Valores de x, y, z")
plt.title("Evolución Temporal")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

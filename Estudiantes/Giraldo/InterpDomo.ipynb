```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

def cargar_imagen(ruta_imagen):
    """Carga y muestra la imagen de fondo."""
    imagen = plt.imread(ruta_imagen)
    return imagen

def graficar_puntos(imagen, x, y):
    """Grafica los puntos seleccionados sobre la imagen."""
    plt.figure(figsize=(10, 6))
    plt.imshow(imagen)
    plt.scatter(x, y, color='red', s=40, label="Datos seleccionados")
    plt.legend()
    plt.axis("on")
    plt.show()

def interpolacion_lagrange(x, y, imagen):
    """Realiza y grafica la interpolación de Lagrange sobre la imagen."""
    polinomio = lagrange(x, y)
    x_nuevo = np.linspace(min(x), max(x), 200)
    y_nuevo = polinomio(x_nuevo)


# Ruta de la imagen (ajustar si es necesario)
ruta_imagen = "planetario-medellin.jpg"
imagen = cargar_imagen(ruta_imagen)

# Definir puntos del domo
# Puntos de interpolación
x_puntos = np.array([360, 348, 350, 386, 409, 433, 456, 486, 514, 538, 569, 595, 622, 666, 689, 710, 725, 737, 744, 745])
y_puntos = np.array([210, 160, 140, 130, 118, 98, 88, 70, 60, 52, 47, 43, 42, 50, 70, 95, 120, 150, 180, 210])

graficar_puntos(imagen, x_puntos, y_puntos)
interpolacion_lagrange(x_puntos, y_puntos, imagen)
```
